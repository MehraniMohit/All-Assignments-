```python
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(6, 6))
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['purple', 'lightcoral'], startangle=90)
plt.title("Distribution of Male and Female Customers")
plt.ylabel('') 
plt.show()
```
